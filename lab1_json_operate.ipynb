```python
# 引入
import os
import json
from pprint import pprint

#=================================================
# 新增一個json
jsonf0 ={
    "number":123,
    "string":"hellocc104",
    "array":[1,2,3,4],
    "object":{"iii":123,"just":"key"}
}
# 列印出jsonobject的內容
pprint(jsonf0)

#=================================================
# 開起一個新檔案testjson
openf = open('./testjson','w')

# 把jsonf0轉成json格式其tpye為str
jsonf1=json.dumps(jsonf0)

# 將jsonf1寫入testjson檔案中
openf.write(jsonf1)

# 將檔案testjson關閉
openf.close()

#=================================================
# 開起testjson檔案
openj = open('./testjson','r')

# 載入json內容至jsonf2變數中
jsonf2 = json.load(openj)

# 將檔案testjson關閉
openj.close()

#=================================================
# 重新指定jsonf2變數中的"a"key value 對應 數字
jsonf2["number"]=321

# 重新指定jsonf2變數中的"b"key value 對應 字串
jsonf2["string"]="heycc104"

# 重新指定jsonf2變數中的"c"key 對應 json number array 
jsonf2["array"]=[4,3,2,1]

# 重新指定"jsonf2變數中的d"key value  對應 json string array
jsonf2["object"]={"iii":"key","just":123}

#=================================================
# 列印出jsonf2的內容
pprint(jsonf2)

# jsonf2的內容轉成json格式其tpye為str並存入jsonf3變數中
jsonf3=json.dumps(jsonf2)

#=================================================
# 開起一個檔案testjson2
openf1 = open('./testjson','w')

# 將jsonf3變數寫入檔案
openf1.write(jsonf3)

# 將檔案關閉
openf1.close()
```
